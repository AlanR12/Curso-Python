Haremos unos cálculos antes de proceder a los incisos, primero definimos: 
$D_1 \sim Bernoulli(q_1)$ & $D_2 \sim Bernoulli(q_2)$, por lo tanto: $S = \sum_{j=1}^{2}C_{i}D_{i}$

Ahora encontraremos los valores de $q_1$ y $q_2$:


Sabemos que $P[S = 0]= .24$, es decir:

$$P[S = 0] = P\left[\sum_{j=1}^2 C_i D_i = 0\right]$$

$$= P[ C_1 D_1 + C_2 D_2 = 0]$$

Lo cual solo sucederá cuando no ocurran reclamaciones, ya que las reclamaciones no pueden tomar valores negativos:

$$= (1 - q_1)(1 - q_2)$$

$$\therefore (1 - q_1)(1 - q_2) = .24 \qquad \qquad \qquad \text{(1)}$$

También sabemos que $P[S = 1] = .1$, es decir:

$$P[S = 1] = P[\sum_{j=1}^2 C_i D_i = 1]$$

$$= P[ C_1 D_1 + C_2 D_2 = 1]$$

Abriremos por casos:

$$= (q_1)(1-q_2)(\frac{1}{6}) + (1-q_1)(q_2)(\frac{1}{4})$$

$$\therefore (q_1)(1-q_2)(\frac{1}{6}) + (1-q_1)(q_2)(\frac{1}{4}) = .1 \qquad \qquad \qquad \text{(2)}$$

Ahora de $(1)$ despejamos $(1-q_1)$:

$$1-q_1 = \frac{.24}{1-q_2}$$

Y sustituímos en $(2)$:

$$\Rightarrow (1-\frac{.24}{1-q_2})(1-q_2)(\frac{1}{6}) + (\frac{.24}{1-q_2})(q_2)(\frac{1}{4}) = .1$$

$$\frac{.24 q_2}{4(1-q_2)} + \frac{1}{6} (1- \frac{.24}{1-q_2})(1-q_2) =\frac{1}{10}$$

$$\frac{3 q_2}{50(1-q_2)} + \frac{1}{6} (1- \frac{6}{25(1-q_2)}) (1-q_2) = \frac{1}{10}$$

$$\frac{3 q_2}{50(1-q_2)} + \frac{1}{6} (\frac{25(1-q_2) - 6}{25(1-q_2)}) (1-q_2) = \frac{1}{10}$$

$$\frac{3 q_2}{50(1-q_2)} + \frac{1}{6} (\frac{25(1-q_2)^2 - 6(1-q_2)}{25(1-q_2)}) = \frac{1}{10}$$

$$\frac{3 q_2}{50(1-q_2)} + \frac{1}{6} (\frac{25 - 50q_2 + 25q_2^2 - 6 + 6q_2}{25(1-q_2)}) = \frac{1}{10}$$

$$\frac{3 q_2}{50(1-q_2)} + \frac{25 - 50q_2 + 25q_2^2 - 6 + 6q_2}{150(1-q_2)} = \frac{1}{10}$$

$$\frac{9q_2 + 25 - 50q_2 + 25q_2^2 - 6 + 6q_2}{150(1-q_2)} = \frac{1}{10}$$

$$\frac{25q_2^2 - 35q_2 + 19}{150(1-q_2)} = \frac{1}{10}$$

$$-25q_2^2 + 35q_2 - 19 - 15q_2 + 15 = 0$$

$$-25q_2^2 + 20q_2 - 4 = 0$$

Que al resolver con fórmula general, nos queda:

$$\therefore q_2 = \frac{2}{5} =.4$$

Si sustituímos $q_2=.4$ en (1):

$$ 1-q_1 = \frac{.24}{1-.4} \Rightarrow q_1 = .6$$

$$\therefore D_1 \sim Bernoulli(q_1=.6) \quad \text{y} \quad D_2 \sim Bernoulli(q_2=.4)$$

````python
## Para ahorrarte la talacha puedes colocar el sistema de ecuaciones de la siguiente manera con Sympy
## El objeto Eq es como tal la ecuación puede ser igualada a 0 o a un valor cualquiera

import numpy as np
import sympy as sy

q_1, q_2 = sy.symbols('q1 q2')

eq1 = sy.Eq((1-q_1)*(1-q_2)-0.24,0)
eq2 = sy.Eq(q1*(1-q_2)*(1/6)+(1-q_1)*q_2*(1/4)-0.1,0)

solucion = sy.solve((eq1, eq2),(q_1, q_2))

q1 = solucion[0][0]
q2 = solucion[0][1]

print("q1 = {:.2f}".format(q1))
print("q2 = {:.2f}".format(q2))
````

# Ejercicio 2

Una aseguradora tiene un portafolio con $100$ pólizas del ramo de autos, las primeras $30$ están sujetas a un deducible de $\$50$, otras $40$ pólizas a un deducible de $\$90$ y el restante a un deducible de $\$25$. El seguro pagará el excedente al deducible con un monto máximo de beneficio de $\$300$.

El número de accidentes de autos para todo el portafolio tiene una distribución $Poisson$ con media 3 y los montos de reclamación tienen la siguiente función de masa de probabilidad.

x  | $150$     $250$     $350$ 
:---: | ------------------- 
*f(x)*| $1/6$   $1/2$   $1/3$   
 
Calcula el monto esperado que tendrá que pagar la aseguradora.

***SOLUCIÓN***

Sea $S = \sum_{i=1}^N X_i$, por lo que $E[S] = E[N]E[X]$ de donde ya conocemos $E[N]=3$. Así que nos enfocaremos en encontrar $E[X]$.

Para esto, primero debemos de reescribir a $X$ en término del deducible y del monto máximo, es decir, como $max\{min\{X,u\}-d,0\}$:

 $\cdot$ Sabemos que como tenemos 30 pólizas de 100 totales con $d=50$, se tiene un $p=.3$ de que al suceder un accidente se trate de una de estas pólizas.

 $\cdot$ Como tenemos 40 pólizas de 100 totales con $d=90$, se tiene un $p=.4$ de que al suceder un accidente se trate de una de estas pólizas.
 
 $\cdot$ Como tenemos 30 pólizas de 100 totales con $d=25$, se tiene un $p=.3$ de que al suceder un accidente se trate de una de estas pólizas.
 
 Todas con un monto máximo de $u=300$.

Entonces podemos reescribir la tabla como:

***x*** *p =0.3*   | $100$     $200$     $250$ 
:---: | ------------------- 
***x*** *p =0.4*   | $60$     $160$     $210$ 
***x*** *p =0.3*   | $125$     $225$     $275$ 
*f(x)*| $1/6$   $1/2$   $1/3$  

Es decir, que la $E[X]$ se verá de la siguiente forma:
$$E[X] = (.3)[100 \cdot \frac{1}{6} + 200 \cdot \frac{1}{2} + 250 \cdot \frac{1}{3}] + (.4)[60 \cdot \frac{1}{6} + 160 \cdot \frac{1}{2} + 210 \cdot \frac{1}{3}] + (.3)[125 \cdot \frac{1}{6} + 225 \cdot \frac{1}{2} + 275 \cdot \frac{1}{3}]$$
$$\therefore E[X] = \frac{383}{2}$$
Sin embargo, se trata de una póliza, por lo que para calcular $E[S]$ hay que multiplicar por 100 pólizas.

Entonces tendremos que:
$$E[S] = 100\cdot E[N]E[X]$$
$$= 100 \cdot 3 \cdot \frac{383}{2} = 57450$$
$$\therefore E[S] = \text{Monto esperado a pagar de la aseguradora} = 57,450$$




````python
# Para confirmar las operaciones: 
import numpy as np
P1 = 0.3 
P2 = 0.4
P3 =0.3
x1 = np.array([100,200,250]) 
x2 = np.array([60,160,210])
x3 = np.array([125,225,275])
f_x = [1/6,1/2,1/3]
E_X = (P1*(x1*f_x))+(P2*(x2*f_x))+(P3*(x3*f_x))
E_X1 = (E_X[0]+E_X[1]+E_X[2])
print("E[X] = ",E_X1)
E_N = 3
E_S = 100*E_N*E_X1

print("E[S]= ",E_S)
````

Para aproximar la función de densidad de S con el método de Gamma trasladada, necesitaremos reescribir la variable encontrar la variable $G= Z + c$ que se ajustará a $S$, donde $Z \sim Gamma(\beta, \lambda)$. Es decir, tenemos que encontrar los valores de $c$, $\beta$ y $\lambda$. Para esto necesitamos $E[S]$ y $Var[S]$.

$\quad \cdot \quad E[S]$
$$E[S] = E[N] E [X] = (10)(1\cdot 1) = 10$$
$$\therefore \mu = 10$$

$\quad \cdot \quad Var[S]$
$$Var[S] = Var[N] E^2[X] + Var[X] E[N]$$

$\quad \cdot \quad E[S^2]$

$$E[S^2] = Var[S] + E^2[S]$$

$\quad \cdot \quad \mbox{Coeficiente de asimetría}=\alpha$

Sabemos que $\alpha = \frac{E[(S-E[S])^3]}{\sqrt{Var^3[S]}}$, es decir:
$$\alpha = \frac{E[S^3 - 3 S^2 E[S] + 3SE^2[S]- E^3[S]]}{\sqrt{Var^3[S]}}$$

Se sabe que $N$ es de la familia $(a,b,0)$. $X_1$ sigue una distribución exponencial con media $\theta$ la media y varianza de $S_1$ son 84 y 2156 respectivamente. $X_2$ sigue una distribución Unif$(0,\theta)$, la media y varianza de $S_2$ son 42 y 343 respectivamente. Calcula $P[N=2]$

***SOLUCIÓN***

Primero, recopilemos los datos: $S_1 = \sum_{j=1}^N X_{1j}$ y $S_2 = \sum_{j=1}^N X_{2j}$, en donde $N$ es clase(a,b,0), $X_1 \sim exp(\frac{1}{\theta})$ y $X_2 \sim Unif(0,\theta)$.

$$
S_1=
\begin{cases}
E[S_1] \hspace{0.8cm} \text{=} \hspace{0.4cm} 84\\
Var[S_1] \hspace{0.8cm} \text{=} \hspace{0.4cm} 2156\\
\end{cases}
$$

$$
S_2=
\begin{cases}
E[S_2] \hspace{0.8cm} \text{=} \hspace{0.4cm} 42\\
Var[S_2] \hspace{0.8cm} \text{=} \hspace{0.4cm} 343\\
\end{cases}
$$

Ahora, sabemos que como se trata de un modelo coletivo tenemos que:
$$E[S]=E[N]E[X]$$
Es decir:

$$ E[S_1]=E[N]E[X_1] = E[N] \theta = 84 \quad \Rightarrow \quad E[N] = \frac{84}{\theta}$$
$$ E[S_2]=E[N]E[X_2] = E[N] \frac{\theta}{2} = 42 \quad \Rightarrow \quad E[N] = \frac{84}{\theta}$$
De la misma forma, tenemos que:
$$Var[S_1] = Var[N]E^2[X_1] + Var[X_1]E[N]$$
$$= Var[N] \theta^2 + \theta^2 \cdot \frac{84}{\theta}$$
$$= Var[N] \theta^2 + 84 \theta = 2156$$
$$\therefore Var[N] = \frac{2156-84\theta}{\theta^2}\quad \quad \quad \quad \quad \quad (1)$$
Por otro lado:
$$Var[S_2] = Var[N]E^2[X_2] + Var[X_2]E[N]$$
$$= Var[N] (\frac{\theta}{2})^2 + \frac{\theta^2}{12} \cdot \frac{84}{\theta}=343$$
Sustituímos $(1)$:
$$\rightarrow \frac{2156-84\theta}{\theta^2} (\frac{\theta}{2})^2 + \frac{\theta^2}{12} \cdot \frac{84}{\theta}=343$$
$$\frac{2156-84\theta}{4} + \frac{84\theta}{12}=343$$
$$\frac{3(2156-84\theta) + 84\theta}{12}=343$$
$$3(2156-84\theta) + 84\theta=343 \cdot 12$$
$$6468 - 252 \theta + 84 \theta = 4116$$
$$2352 = 168 \theta$$
$$\therefore \theta = \frac{2352}{168} = 14$$
Sustituímos el valor de $\theta$ $E[N]$ y $Var[N]$:
$$E[N]=\frac{84}{14} = 6$$
y
$$Var[N] = \frac{2156-84(14)}{14^2} = 5$$
Es decir: $E[N] > Var[N]$
$$\therefore N \sim Bin(n,p)$$
Por lo tanto, podemos encontrar los parámetros de la siguiente forma:
$$E[N] = 6 = np \quad \quad \quad \quad \quad \quad  (\cdot)$$
$$Var[N] = 5 = np(1-p) \quad \quad \quad \quad \quad \quad  (\cdot \cdot)$$
Sustituimos $(\cdot)$ en $(\cdot \cdot)$:
$$6(1-p) = 5$$
$$1-p = \frac{5}{6}$$
$$\therefore p = \frac{1}{6}$$
Sustituimos $p$ en $(\cdot)$:
$$6 = n(\frac{1}{6})$$
$$\therefore n = 36$$
Entonces podemos decir que:
$$P[N=2] = \binom{36}{2} \cdot (\frac{1}{6})^2 (1-\frac{1}{6})^{36-2}$$
$$\therefore P[N=2] \approx.035553$$


````python
# Corrobación de las operaciones 
import sympy
theta = 2352/168
print("θ = ",theta)

#E[N]
E_N = 84/14
print("E[N] = ", E_N)
#Var[N]
Var_N = (2156-(84*14))/14**2
print("Var[N] = ", Var_N)

#Combianciones
comb = binomial(36, 2)
print("# Combinaciones:",comb)

#P[N=2]
P_N_2 = comb*((1/6)**2)*((1-1/6)**(36-2))
print("P[N=2] = ", P_N_2)
````

# Ejercicio 5
$S$ tiene una distribución compuesta con frecuencia $N$ y severidad $X$, $N$ sigue una distribuión Poisson de media 1, y $X$ tiene función de probabilidad:

$$\mathbb{P}\left[X=j\right]=\frac{\beta^j}{(1+\beta)^{j+1}} \hspace{1.5cm} \forall j\in \mathbb{N}\cup \{0\}$$

Si $\mathbb{E}\left[S|S>0\right]=\frac{1}{2-2e^{-1/3}}$. Determina $\mathbb{E}\left(X\right)$.

***SOLUCIÓN***  

$$\Rightarrow  \mathbb{E}[S | S>0] = \frac{1}{2-2e^{-\frac{1}{3}}}= \frac{\mathbb{E}[S]}{\mathbb{P}[S>0]} = \frac{\mathbb{E}[N]\mathbb{E}[X]}{1-\mathbb{P}[S=0]}=\frac{\mathbb{E}[X]}{1-\mathbb{P}[S=0]}$$  
Este último paso debido a que $\mathbb{E}[N]=\frac{1}{1}=1$  
Entonces para $\mathbb{P}[S=0]$:
$$\Rightarrow \mathbb{P}[S=0]=\mathbb{E}[(\mathbb{P}[X=0])^N]=M_N(\mathbb{P}[X=0])$$
$$\hspace{2cm} =e^{-\lambda(1-\mathbb{P}[X=0])}$$
donde $\mathbb{P}[X=0]=\frac{1}{1+\beta}, \lambda = 1$:
$$\hspace{2cm} =e^{-(1-(1-\beta)^{-1})} $$
$$\hspace{2cm} =e^{-(\frac{1+\beta-1}{1+\beta})} = e^{-\frac{\beta}{1+\beta}}$$
$$ \therefore \mathbb{E}[S|S>0]=\frac{\mathbb{E}[X]}{1-e^-\frac{\beta}{1+\beta}}=\frac{1}{2-2e^{-\frac{1}{3}}}$$
$$ \therefore \mathbb{E}[X]=\frac{1-e^{-\frac{\beta}{1+\beta}}}{2-2e^{-\frac{1}{3}}}$$

Por otro lado, tenemos que
$$\mathbb{P}[X=j]=\frac{\beta^j}{(1+\beta)^{j+1}}$$
$$\Rightarrow \mathbb{E}[X]=\sum _{j=0}^{\infty }\:jf_x(j)=\sum _{j=0}^{\infty }\:j\frac{\beta^j}{(1+\beta)^{j+1}}$$
$$= \sum _{j=0}^{\infty }\:\frac{j}{1+\beta}*(\frac{\beta}{1+\beta})^j$$
$$=\frac{1}{1+\beta}\sum _{j=0}^{\infty }\:j(\frac{\beta}{1+\beta})^j \hspace{1.5cm} tomemos \quad \frac{\beta}{1+\beta}=a$$
$$\Rightarrow \mathbb{E}[X]=\frac{1}{1+\beta}\sum _{j=0}^{\infty }\:ja^j$$
Sabemos que $1+\beta > \beta \Rightarrow |a|<1$
Por definición de la serie que tenemos (que es serie de potencias)
$$= \frac{1}{1+\beta}(\frac{a}{(a-1)^2}) = \frac{1}{1+\beta}(\frac{a}{a^2-2a+1})$$
$$= \frac{1}{1+\beta}(\frac{\frac{\beta}{1+\beta}}{\frac{\beta^2}{(1+\beta)^2}-2\frac{\beta}{1+\beta}+1})$$
$$= \frac{1}{1+\beta}(\frac{\frac{\beta}{1+\beta}}{\frac{\beta^2-2\beta(1+\beta)+(1+\beta)^2}{(1+\beta)^2}})$$
$$= \frac{1}{1+\beta}(\frac{\frac{\beta}{1+\beta}}{\frac{\beta^2-2\beta(1+\beta)+(1+\beta)^2}{(1+\beta)^2}})$$
$$= \frac{1}{1+\beta}(\frac{\beta(1+\beta)}{\beta^2-2\beta-2\beta^2+1+2\beta+\beta^2})$$
$$= \beta$$
$$\therefore \mathbb{E}[X]=\beta ...(1) \hspace{.5cm} y \hspace{.5cm} \mathbb{E}[X]=\frac{1-e^{-\frac{\beta}{1+\beta}}}{2-2e^{-\frac{1}{3}}} ...(2) $$
$$\Rightarrow \beta=\frac{1-e^{-\frac{\beta}{1+\beta}}}{2-2e^{-\frac{1}{3}}}$$
$$\Rightarrow \beta(2-2e^{-\frac{1}{3}})=1-e^{-\frac{\beta}{1+\beta}}$$
$$\Rightarrow \beta(2-2e^{-\frac{1}{3}})+e^{-\frac{\beta}{1+\beta}}=1$$
$$\Rightarrow f(\beta)=1$$
Vemos que para que se cumpla la ecuación anterior, entonces $\beta=\frac{1}{2}$ o $\beta=0$, para ver esto, nos apoyaremos del código en R y obtenemos la siguiente gráfica:

```{r}
curve((2-(2*exp(-1/3)))*x,col="coral3", lwd=2,from=0, to=.7, xlab= "beta", ylab="y")
curve(1-(exp(-x/(x+1))), from=0, to=.7, add = TRUE ,xlab="beta", ylab="y", col="aquamarine4", lwd=2)
grid()
```



Con esta gráfica podemos observar que nuestras gráficas tienen dos intersecciones, cuando $\beta=0$ y $\beta=\frac{1}{2}$, así que ahora sustituiremos ambos valores en estas ecuaciones para ver si se cumple la igualdad
Si  $$ \beta=0  \Rightarrow 0(2-2e^{-\frac{1}{3}})+e^{-\frac{0}{1+0}} = 0 + e^0 = 1 $$
$$\therefore f(\beta)=1 \hspace{1cm} \text{ con  } \beta=0$$
Si  $$ \beta=\frac{1}{2} \Rightarrow \frac{1}{2}(2-2e^{-\frac{1}{3}})+ e^{-\frac{1}{3}} = 1 - e^{-\frac{1}{3}} +e^{-\frac{1}{3}} = 1 $$

Vemos que con ambos valores obtenemos que se cumple la igualdad $f(\beta)=1$, por lo que entonces para $\beta$ tenemos los valores de $\beta_1= 0$ y $\beta_2=\frac{1}{2}$

Al final, tomamos el valor de $\beta = \frac{1}{2}$ ya que si tomamos el valor de $\beta =0$, entonces para cual quier $\forall j \in \mathbb{N}\text{U{0}}, \mathbb{P}[X=j]=0,  \text{ ya que } =\frac{\beta^j}{(1+\beta)^{j+1}}=\frac{0^j}{1^{j+1}}=0$
$$\therefore \mathbb{E}[X] = \frac{1}{2}$$
